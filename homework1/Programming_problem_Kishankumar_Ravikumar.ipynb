```python
#
#   Question 1
#  1) Please plot the noisy data and the polynomial you found (in the same figure). 
#  You can use any value of m selected from 2, 3, 4, 5, 6.

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
noise_scale = 100
number_of_samples = 50
x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 7 * x + 15 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
plt.show()
```

```python
# Here we have chosen m = 3

x = x.ravel()
y = y.ravel()
z = np.polyfit(x, y, 3)
p = np.poly1d(z)
plt.plot(x, y, 'ro', label = 'Noisy Data')
plt.plot(x,p(x),'bo',label = 'Polynomial')
plt.legend()
plt.show()
#reference - https://docs.scipy.org/doc/numpy/reference/generated/numpy.polyfit.html
```

```python
#Q2. Plot MSE versus order m for m = 1, 2, 3, 4, 5, 6, 7, 8. 
#    Identify the best choice of order m.


x = x.ravel()
y = y.ravel()
m = []
for i in [1, 2, 3, 4, 5, 6, 7, 8]:
    z = np.polyfit(x, y, i)
    p = np.poly1d(z)
    m.append(((sum((y-p(x))**2)))/number_of_samples)
    plt.title(f'M = {i}')
    plt.plot(x, y, 'ro', label = 'ND ') #ND = NOISY DATA 
    plt.plot(x,p(x),'bo',label = 'Poly') # ( POLY = POLYNOMIAL) 
    plt.legend()
    plt.show()
print("MSE: ",m)
plt.plot([1, 2, 3, 4, 5, 6, 7, 8],m)
plt.xlabel('M')
plt.ylabel('MSE')
plt.show()
```

ANSWER FOR QUESTION 2 =
    As per result from thegraph plots above the
best value of m would be equal to 3.


```python
#Q3. Change variable noise_scale to 150, 200, 400, 600, 1000 respectively, re-run the algorithm and
#    plot the polynomials with the m found in 2). Discuss the impact of noise scale to the accuracy of
#    the returned parameters. You need to plot a figure like in 1) for each choice of noise_scale.

for noise_scale in [150, 200, 400, 600, 1000]: #n noise  scale range
    number_of_samples = 50
    x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
    y = 7 * x + 15 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1) 
    x = x.ravel()
    y = y.ravel()
    z = np.polyfit(x, y, 3)
    p = np.poly1d(z)
    plt.title(f'noise_scale = {noise_scale}')
    plt.plot(x, y, 'ro', label = 'Noisy Data')
    plt.plot(x,p(x),'bo',label = 'Polynomial')
    plt.legend()
    plt.show()
```
